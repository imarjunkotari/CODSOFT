# Relation of Heartrate and Minutely Intensity

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure Id columns are the same type (string)
heartrate['Id'] = heartrate['Id'].astype(str)
hourlyintensity['Id'] = hourlyintensity['Id'].astype(str)

# Align time columns to the hour
heartrate['Hour'] = pd.to_datetime(heartrate['Time']).dt.floor('H')
hourlyintensity['Hour'] = pd.to_datetime(hourlyintensity['ActivityHour']).dt.floor('H')

# Aggregate heart rate
avg_heartrate = heartrate.groupby(['Id', 'Hour'])['Value'].mean().reset_index()

# Merge DataFrames
merged = pd.merge(avg_heartrate, hourlyintensity, on=['Id', 'Hour'])

# Verify columns exist
print("Merged columns:", merged.columns.tolist())

# Plot
plt.figure(figsize=(12,6))
sns.regplot(x='Value', y='TotalIntensity', data=merged, scatter_kws={'alpha':0.5}, line_kws={'color':'pink'})
plt.xlabel('Average Heart Rate (per hour)')
plt.ylabel('Total Intensity')
plt.title('Heart Rate vs. Minute Intensity')
plt.tight_layout()
plt.show()
```
